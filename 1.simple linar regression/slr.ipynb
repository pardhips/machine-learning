# simple linear regression
1 independent feature , 1 target feature

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
```

```python
df = pd.read_csv("50_Startups.csv")
df.head()
```

# estimate profit based on rnd budget 

# perform scatterplot on profit vs rnd

```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot( data=df, x="RND", y="PROFIT")
plt.title("PROFIT VS RND")
plt.show()
```

# seperate x and y


```python
x = df[["RND"]]
y = df[["PROFIT"]]
```

```python
x.head()
```

```python
y.head()
```

```python
x.values
```

# create a linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_
intercept = model.intercept_
print(f"slope:  {slope},  intercept:  {intercept}")
```

# ycap = B0 + B1*x

 profit(est) = 49032.89 + 0.8542*RND

# out of sample prediction
estimate the profit for startup with RND spend os 100000

```python
RND = 1_00_000 #1,00,000
pred_profit = model.predict([[RND]])
print(f"predicted profit:  {pred_profit}")
```

```python
plt.figure(figsize = (10,5))
sns.regplot(data=df, x="RND", y="PROFIT" , line_kws={"color":"red"})
plt.title("REGRESSION PLOT PROFIT VS RND")
plt.grid()
plt.show()
```
